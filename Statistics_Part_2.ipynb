# Theory Question


1  What is hypothesis testing in statistics?

Answer- Hypothesis testing, sometimes called significance testing, is an act in statistics whereby an analyst tests an assumption regarding a population parameter.


2  What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer- The null hypothesis of a test always predicts no effect or no relationship between variables, while the alternative hypothesis states your research prediction of an effect or relationship.

3 What is the significance level in hypothesis testing, and why is it important?

Answer- the probability of rejecting the null hypothesis when it is actually true, representing the risk of making a Type I error. It's crucial because it sets the threshold for determining statistical significance, guiding researchers in deciding whether to reject or fail to reject the null hypothesis.

Why it's important:


Controls Type I Error: By setting a significance level, researchers can control the probability of making a Type I error (false positive), which is incorrectly rejecting the null hypothesis when it is true.


Determines Statistical Significance: The significance level is used to determine whether the results of a statistical test are statistically significant. If the p-value (the probability of observing the data, assuming the null hypothesis is true) is less than or equal to the significance level, the null hypothesis is rejected.


Guides Decision-Making: The significance level helps researchers make informed decisions about the validity of their hypotheses and the conclusions they draw from their data.


Standard in Research: A commonly used significance level is 0.05, which means there's a 5% chance of incorrectly rejecting the null hypothesis when it's true.

4   What does a P-value represent in hypothesis testing?

Answer- In statistical hypothesis testing, P-Value or probability value can be defined as the measure of the probability that a real-valued test statistic is at least as extreme as the value actually obtained. P-value shows how likely it is that your set of observations could have occurred under the null hypothesis. P-Values are used in statistical hypothesis testing to determine whether to reject the null hypothesis. The smaller the p-value, the stronger the likelihood that you should reject the null hypothesis.



5  How do you interpret the P-value in hypothesis testing?

Answer- A p-value less than 0.05 is typically considered to be statistically significant, in which case the null hypothesis should be rejected. A p-value greater than 0.05 means that deviation from the null hypothesis is not statistically significant, and the null hypothesis is not rejected.


6  What are Type 1 and Type 2 errors in hypothesis testing?

Answer- In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, while a Type II error (false negative) occurs when you fail to reject a false null hypothesis.

Here's a more detailed explanation:


Type I Error (False Positive):


This happens when you conclude that there is a significant effect or difference when, in reality, there is none.


You incorrectly reject the null hypothesis (which is actually true).
Think of it as concluding someone is guilty when they are actually innocent.


Type II Error (False Negative):


This happens when you fail to detect a real effect or difference that actually exists.


You incorrectly fail to reject the null hypothesis (which is actually false).
Think of it as concluding someone is innocent when they are actually guilty.

7  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Answer- a one-tailed test focuses on detecting a difference in one specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (greater than or less than).

Here's a more detailed breakdown:

let’s discuss the meaning of a one-tailed test.  If you are using a significance level of .05, a one-tailed test allots all of your alpha to testing the statistical significance in the one direction of interest.  This means that .05 is in one tail of the distribution of your test statistic. When using a one-tailed test, you are testing for the possibility of the relationship in one direction and completely disregarding the possibility of a relationship in the other direction.  Let’s return to our example comparing the mean of a sample to a given value x using a t-test.  Our null hypothesis is that the mean is equal to x. A one-tailed test will test either if the mean is significantly greater than x or if the mean is significantly less than x, but not both. Then, depending on the chosen tail, the mean is significantly greater than or less than x if the test statistic is in the top 5% of its probability distribution or bottom 5% of its probability distribution, resulting in a p-value less than 0.05.  The one-tailed test provides more power to detect an effect in one direction by not testing the effect in the other direction.

let’s start with the meaning of a two-tailed test.  If you are using a significance level of 0.05, a two-tailed test allots half of your alpha to testing the statistical significance in one direction and half of your alpha to testing statistical significance in the other direction.  This means that .025 is in each tail of the distribution of your test statistic. When using a two-tailed test, regardless of the direction of the relationship you hypothesize, you are testing for the possibility of the relationship in both directions.  For example, we may wish to compare the mean of a sample to a given value x using a t-test.  Our null hypothesis is that the mean is equal to x. A two-tailed test will test both if the mean is significantly greater than x and if the mean significantly less than x. The mean is considered significantly different from x if the test statistic is in the top 2.5% or bottom 2.5% of its probability distribution, resulting in a p-value less than 0.05.     

8   What is the Z-test, and when is it used in hypothesis testing?

Answer- A Z-test is a statistical test used to determine if there's a significant difference between a sample mean and a population mean or between the means of two populations when the population standard deviation is known or the sample size is large, typically greater than 30.



9  How do you calculate the Z-score, and what does it represent in hypothesis testing?

Answer- A Z-score quantifies how many standard deviations a data point is from the mean of a distribution. In hypothesis testing, Z-scores are used to determine the probability of observing a sample statistic as extreme or more extreme than the one observed, assuming the null hypothesis is true. This probability, known as the p-value, helps decide whether to reject or fail to reject the null hypothesis.


10  What is the T-distribution, and when should it be used instead of the normal distribution?

Answer-  The t-distribution, also known as Student's t-distribution, is a probability distribution similar to the normal distribution but with heavier tails, used when estimating population parameters from small samples or when the population standard deviation is unknown. You should use the t-distribution instead of the normal distribution when the sample size is small (typically less than 30) and the population standard deviation is unknown.


11  What is the difference between a Z-test and a T-test?

Answer- The key difference between a Z-test and a T-test lies in the sample size and knowledge of the population standard deviation:

Z-test:


Used when the population standard deviation is known, or when the sample size is large (generally 30 or more). It assumes the population follows a normal distribution.

T-test:


Used when the population standard deviation is unknown, particularly when the sample size is small (generally less than 30). It also assumes the population is normally distributed, but accounts for the uncertainty in estimating the population standard deviation from a small sample.

12  What is the T-test, and how is it used in hypothesis testing?

13  What is the relationship between Z-test and T-test in hypothesis testing?

14  What is a confidence interval, and how is it used to interpret statistical results?

It quantifies the uncertainty around a sample estimate and provides a level of confidence that the true population parameter falls within that range. For example, a 95% confidence interval suggests that if the sampling process were repeated many times, 95% of the resulting intervals would contain the true population mean.

15  What is the margin of error, and how does it affect the confidence interval?

Answer-  The margin of error represents the potential inaccuracy of a sample statistic as an estimate of a population parameter, while a confidence interval provides a range within which the true population parameter is likely to fall. The margin of error is half the width of the confidence interval.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

# Z-test: Compare sample mean to known population mean
from scipy import stats
import numpy as np

# Sample data
sample = np.array([45, 50, 55, 60, 65, 70, 75, 80])
pop_mean = 60
pop_std = 10
n = len(sample)

# Z-test calculation
z_score = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Results
print(f"Z-score: {z_score:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

# Function to visualize Type I and Type II errors
def visualize_errors(mu0, mu1, sigma, n, alpha):
    import matplotlib.pyplot as plt
    from scipy.stats import norm

    se = sigma / np.sqrt(n)
    critical_value = norm.ppf(1 - alpha/2)
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0')
    plt.plot(x, y1, label='H1')
    cv_low = mu0 - critical_value * se
    cv_high = mu0 + critical_value * se
    plt.axvline(cv_low, color='red', linestyle='--')
    plt.axvline(cv_high, color='red', linestyle='--')
    plt.fill_between(x, 0, y0, where=(x < cv_low) | (x > cv_high), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, y1, where=(x >= cv_low) & (x <= cv_high), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.show()

visualize_errors(100, 105, 15, 50, 0.05)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

# Confidence interval function
def confidence_interval(data, confidence=0.95):
    import scipy.stats as stats
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    margin = stderr * stats.t.ppf((1 + confidence) / 2, n - 1)
    return (mean - margin, mean + margin)

sample = np.random.normal(100, 10, 30)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

# Bayesian Inference using Bayes' Theorem
# P(A|B) = P(B|A) * P(A) / P(B)
def bayes_theorem(p_a, p_b_given_a, p_b):
    return (p_b_given_a * p_a) / p_b

# Example
p_a = 0.01  # Probability of disease
p_b_given_a = 0.9  # Probability of positive test given disease
p_b = 0.05  # Total probability of positive test

posterior = bayes_theorem(p_a, p_b_given_a, p_b)
print(f"Posterior Probability (P(Disease|Positive Test)): {posterior:.4f}")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python


import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: contingency table
data = [[30, 10], [20, 40]]
df = pd.DataFrame(data, columns=['Category A', 'Category B'], index=['Group 1', 'Group 2'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(df)

print("Chi-square Statistic:", round(chi2, 4))
print("P-value:", round(p, 4))
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

if p < 0.05:
    print("Reject the null hypothesis: Variables are dependent")
else:
    print("Fail to reject the null hypothesis: Variables are independent")
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples

# F-test to compare variances
import numpy as np
from scipy import stats  # Import the stats module

# F-test to compare variances
np.random.seed(1)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(52, 15, 30)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

f_stat = var1 / var2
df1 = len(group1) - 1
df2 = len(group2) - 1

# Use the larger variance as numerator to ensure F >= 1
if var1 > var2:
    f_stat = var1 / var2
    df1, df2 = df1, df2
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)
else:
    f_stat = var2 / var1
    df1, df2 = df2, df1
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

# ANOVA test
from scipy.stats import f_oneway

group1 = np.random.normal(70, 5, 30)
group2 = np.random.normal(72, 5, 30)
group3 = np.random.normal(75, 5, 30)

f_stat, p_val = f_oneway(group1, group2, group3)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Reject the null hypothesis: At least one group mean is different")
else:
    print("Fail to reject the null hypothesis: All group means are similar")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

# ANOVA assumptions check
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    print("Shapiro-Wilk Test for Normality:")
    for i, g in enumerate(groups):
        stat, p = shapiro(g)
        print(f" Group {i+1}: W={stat:.4f}, P={p:.4f}")

    stat, p = levene(*groups)
    print(f"\nLevene’s Test for Equal Variance: W={stat:.4f}, P={p:.4f}")

check_anova_assumptions(group1, group2, group3)
```
